```python
import pandas as pd
```

```python
players = pd.read_csv('data/interim/all_player_stats_00-24.csv')
all_stars = pd.read_csv('data/interim/allstar_stats_00-24.csv')
```

```python
all_stars
```

```python
all_stars = all_stars[['Player', 'Season']]
all_stars['All Star'] = True
all_stars
```

```python
df = pd.merge(players, all_stars, on=['Player', 'Season'], how='outer')
```

```python
df['All Star'] = df['All Star'].fillna(False)
```
